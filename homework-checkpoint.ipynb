```python
#Pandas-challange
#week-4
#coding by Prathima_Polavarapu

# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#calculate the total number of schools

total_schools=len(school_data_complete["school_name"].unique())
#print(total_schools)


#calculate the total number of students

total_students = school_data_complete["student_name"].count()
#print(total_students)

#calculate the total budget

total_budget = school_data_complete["budget"].unique().sum()
#print(total_budget)

#calculating average math

average_math = school_data_complete["maths_score"].mean()
#print(average_math)

#calculating average reading

average_reading = school_data_complete["reading_score"].mean()
#print(average_reading)

#Calculate the percentage of students with a passing maths score (50 or greater)

math_passing_score = school_data_complete.loc[(school_data_complete["maths_score"]>=50)]
stu_pass_math=math_passing_score["maths_score"].count()

percent_math = (stu_pass_math/total_students) * 100
#print(math_passing_score)
#percent_math

#Calculate the percentage of students with a passing reading score (50 or greater)

reading_passing_score = school_data_complete.loc[(school_data_complete["reading_score"]>=50)]
stu_pass_reading=reading_passing_score["reading_score"].count()

percent_reading = (stu_pass_reading /total_students) * 100
#print(math_passing_score)
#percent_reading

#Calculate the percentage of students who passed maths and reading (% Overall Passing)
passing_df=school_data_complete.loc[(school_data_complete["maths_score"]>=50)
                                    &(school_data_complete["reading_score"]>=50),:]
passing_count=passing_df["reading_score"].count()
overall_passing=passing_count/total_students*100
#overall_passing




```

```python
# Create a dataframe to hold the above results

government_area_summary=pd.DataFrame({"Total Schools":[total_schools],
                                      "Total Students":[total_students],
                                      "Total Budget":[total_budget],
                                      " Average Math Score":[average_math],
                                      "average_reading":[average_reading],
                                      "%passing maths" :[percent_math],
                                      "%passing Reading":[percent_reading],
                                      "%overall passing":[overall_passing]})

government_area_summary
```

* Create an overview table that summarises key metrics about each school, including:

```python
#2. school summary

#School Name
school_name=school_data_complete["school_name"].unique()
school_name

#School Type
schools =school_data_complete.groupby(["school_name"])
school_type=schools["type"].unique()
school_type=school_type.str[0]
#school_type

#Total Students

count_student= schools["reading_score"].count()
#count_student

#Total School Budget
school_budget=schools["budget"].unique()
school_budget=school_budget.str[0]
#school_budget

#Per Student Budget
per_student_budget=school_budget/count_student
#per_student_budget

#Average Maths Score
avg_math_score=schools["maths_score"].sum()/count_student
#avg_math_score

#Average Reading Score
avg_reading_score=schools["reading_score"].sum()/count_student
#avg_reading_score

#% Passing Maths
math_passing=math_passing_score.groupby(["school_name"])
per_passing_math=math_passing["maths_score"].sum()/count_student
#per_passing_math

#% Passing Reading
reading_passing=reading_passing_score.groupby(["school_name"])
per_passing_reading=reading_passing["reading_score"].sum()/count_student
per_passing_reading

#% Overall Passing (The percentage of students that passed maths and reading.)
passing_df=school_data_complete.loc[(school_data_complete["maths_score"]>=50)&(school_data_complete["reading_score"]>=50),:]
passing_gp=passing_df.groupby(["school_name"])
passing_overall=passing_gp["reading_score"].count()/count_student*100
#passing_overall



#Create a dataframe to hold the above results

```
